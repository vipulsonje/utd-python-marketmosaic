### Create a parent Class "Index" (Instruction 1 to 3)
**Class Index includes 8 instance variables, and the instruction doesn't require hidding those instances.
Therefore, we define each instance variables with the first parameter "self".**

```python
class Index:
    def __init__(self, date, index_name, company, type, rating, price, fluct_rate):
        self.date = date
        self.index_name = index_name
        self.company = company
        self.type = type
        self.rating = rating
        self.price = price
        self.fluct_rate = fluct_rate
        self.unique_id = id(self)
# The Index Class has 8 accesssor methods.
    def get_date(self):
        return self.date
    def get_index_name(self):
        return self.index_name
    def get_company(self):
        return self.company
    def get_index_type(self):
        return self.index_type
    def get_rating(self):
        return self.rating
    def get_price(self):
        return self.price
    def get_fluct_rate(self):
        return self.fluct_rate
    def get_unique_id(self):
        return self.unique_id
# The Index Class has 8 mutator methods.
    def set_date(self, date):
        self.date = date
    def set_index_name(self, index_name):
        self.index_name = index_name
    def set_company(self, company):
        self.company = company
    def set_type(self, type):
        self.type = type
    def set_rating(self, rating):
        self.rating = rating
    def set_fluct_rate(self, fluct_rate):
        self.fluct_rate = fluct_rate
    def set_unique_id(self, id):
        self.unique_id = id
# Define __str__() to return a string that content all contents of the object that we are interested in.
    def __str__(self):
        return str(self.unique_id)+','+str(self.date)+',' +str(self.index_name)+','+str(self.company)+',' +str(self.type)+',' +str(self.rating)+',' + str(self.price)+','+str(self.fluct_rate)
```

### Define two subclasses, namely 'LargeCap' and'NonLargeCap'(Instruction step 4 to step 5)
**Both subclasses are inheriting from the 'Index' class.**

```python
class LargeCapIndex(Index):
    def __init__(self, date, index_name, company, rating, price, fluct_rate):
        super().__init__(date, index_name, company, 'LargeCap', rating, price, fluct_rate)
class NonLargeCapIndex(Index):
    def __init__(self, date, index_name, company, rating, price, fluct_rate):
        super().__init__(date, index_name, company,"non-LargeCap", rating, price, fluct_rate)
```

### Define three subclasses, 'SNP','DJ','NASDAQ' (Instruction step 6 to step 7)
**Subclass SNP inherites from Class LargeCapIndex.**

**Subclass DJ and Subclass NASDAQ" inherites from Class NonLargeCapIndex.**

```python
class SNP(LargeCapIndex):
    def __init__(self, date, index_name, rating, price, fluct_rate):
        super().__init__(date, index_name, "SNP", rating, price, fluct_rate)
class DJ(NonLargeCapIndex):
    def __init__(self, date, index_name, rating, price, fluct_rate):
        super().__init__(date, index_name, "DJ", rating, price, fluct_rate)
class NASDAQ(NonLargeCapIndex):
    def __init__(self, date, index_name, rating, price, fluct_rate):
        super().__init__(date, index_name, "NASDAQ", rating, price, fluct_rate)
```

**Average rating for DJ over a year (month wise split)**

```python
monthValues = ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']

list_of_months = []
months = []

from datetime import datetime

for date in project_data['date']:
    date_object = datetime.strptime(date, '%Y-%m-%d') 
    month = date_object.month
    list_of_months.append(month)
#print(list_of_months)

for month in list_of_months:
    if month == 1:
        months.append('January')
    elif month == 2:
        months.append('February')
    elif month == 3:
        months.append('March')
    elif month == 4:
        months.append('April')
    elif month == 5:
        months.append('May')
    elif month == 6:
        months.append('June')
    elif month == 7:
        months.append('July')
    elif month == 8:
        months.append('August')
    elif month == 9:
        months.append('September')
    elif month == 10:
        months.append('October')
    elif month == 11:
        months.append('November')
    else:
        months.append('December')
#print(months)

project_data[month] = months
#print(project_data[month])
```

```python
SNP_DF = project_data.loc[project_data['company'] == 'SNP']
DJ_DF = project_data.loc[project_data['company'] == 'DJ']
NASDAQ_DF = project_data.loc[project_data['company'] == 'NASDAQ']

#print(DJ_DF)
```
